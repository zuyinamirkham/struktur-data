stack dengan list

```python

stack = [] # list kosong bernama stack
# append digunakan untuk menambahkan
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')
print('Initial stack')
print(stack)
# pop untuk menghapus elemen terakhir
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())
print('\nStack after elements are popped:')
print(stack)
# jika kita terus memanggil pop() sampai stack kosong, lalu mencoba pop() lagi, akan muncul error
```

stack dengan collections.deque

```python

from collections import deque
stack = deque() # variabel stack yang berisi deque
# append digunakan untuk menambahkan
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')
print('Initial stack:')
print(stack)
# pop untuk menghapus elemen terakhir
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())
print('\nStack after elements are popped: ')
print(stack)
# jika kita terus memanggil pop() sampai stack kosong, lalu mencoba pop() lagi, akan muncul error
```

stack dengan queue.lifoqueue

```python

from queue import LifoQueue

stack = LifoQueue(maxsize=5) #Membuat objek stack dengan kapasitas maksimal 5 elemen

print(stack.qsize()) # untuk mengetahui jumlah elemen
# put() digunakan untuk menambahkan elemen (operasi push)
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')
print("Full: ", stack.full()) # mengecek elemen apakah sudah full
print("Size: ", stack.qsize()) # menampilkan jumlah elemen
#get() untuk menghapus elemen terakhir
print('\nElements popped from the stack')
print(stack.get())
print(stack.get())
print(stack.get())
print("\nEmpty: ", stack.empty()) # mengecek apakah elemen dalam stack kosong
```

stack dengan single linked list

```python
class Node: # membuat class node pada linked list
  def __init__(self, value): # menginisialisasi node baru
    self.value = value # Menyimpan data/nilai di dalam node
    self.next = None # Pointer yang menunjuk ke node berikutnya (awalnya kosong)
class Stack:
# Saat menginisialisasi (membuat) stack, digunakan dummy node agar lebih mudah menangani kondisi khusus (edge cases).
  def __init__(self):
    self.head = Node("head") # Membuat dummy node (node kosong awal).
    self.size = 0 # menyimpan elemen dalam stack
# bagaimana isi stack ditampilkan dalam bentuk teks saat dicetak.
  def __str__(self):
    cur = self.head.next # mulai dari node pertama(setelah dummy)
    out = ""
    while cur: # perulangan selama masih ada node
      out += str(cur.value) + "->" # menambahakan nilai ke string
      cur = cur.next
    return out[:-2]

  def getSize(self): # mengembalikan jumlah elemen stack
    return self.size

  def isEmpty(self): # mengecek apakah stack kosong
    return self.size == 0

  def peek(self): # melihat elemen teratas tanpa menghapusnya

    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.value

  def push(self, value): # menambahkan elemen baru
    node = Node(value) # membuat node baru
    node.next = self.head.next # node baru menunjuk ke node berikutnya
    self.head.next = node # head menunjuk node baru
    self.size += 1 # menambah ukuran

  def pop(self): # menghapus elemen teratas/terakhir masuk
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

if __name__ == "__main__": # program utama
  stack = Stack() # membuat variabel baru berisi stack
  for i in range(1, 11): # menambahkan elemen 1-10
    stack.push(i)
  print(f"Stack: {stack}")
  for _ in range(1, 6): # menghapus 5 elemen teratas
    remove = stack.pop()
    print(f"Pop: {remove}") # menampilkan elemen yang di hapus
  print(f"Stack: {stack}") # sisa stack
```

tugas

queue dengan list

```python
queue = []

# enqueue
queue.append(10)
queue.append(20)
queue.append(30)

# dequeue
queue.pop(0)

print(queue)
```

queue dengan collection.deque

```python
from collections import deque

queue = deque()

queue.append(10)
queue.append(20)
queue.append(30)

queue.popleft()

print(queue)
```

queue dengan queue.lifoqueue

```python
from queue import Queue

qu = Queue()

qu.put(10)
qu.put(20)
qu.put(30)

print(qu.get())
```
