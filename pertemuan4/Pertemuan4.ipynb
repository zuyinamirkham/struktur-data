queue dengan list

```python

queue = [] # Membuat list kosong yang akan berfungsi sebagai wadah antrean (queue)

# Menambahkan elemen ke dalam antrean (posisi paling belakang)
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')
print('Initial queue')
print(queue) # Mencetak isi list queue saat ini
print('\nElements dequeued from queue:')
print(queue.pop(0)) # mengambil elemen terakhir, bukan yang pertama.
print(queue.pop(0))
print(queue.pop(0))
print('\nQueue after removing elements')
print(queue) # Mencetak sisa isi list queue

# Catatan: Jika memanggil pop() pada list kosong, Python akan melempar error 'IndexError'
```

queue dengan collections.deque

```python
# Mengimpor kelas deque dari modul collections untuk struktur data queue yang efisien
from collections import deque
queue = deque() # Menginisialisasi objek deque kosong sebagai wadah antrean

# Menambahkan elemen belakang antrean
queue.append('a')
queue.append('i')
queue.append('u')
queue.append('e')
queue.append('o')
print('Initial queue:')
print(queue) # Mencetak isi deque

print("\nElements dequeued from the queue")
# Menghapus dan mengembalikan elemen paling depan. Inilah prinsip FIFO yang benar.
print(queue.popleft())
print(queue.popleft())
print(queue.popleft())
print("\nQueue after removing elements")
print(queue)
# Catatan: Jika queue kosong dan popleft() dipanggil, akan muncul IndexError.
```

queue dengan queue.queue

```python
# Mengimpor kelas Queue dari modul bawaan Python 'queue'
from queue import Queue
# Membuat objek antrean (qu) dengan kapasitas maksimal 5 elemen
qu = Queue(maxsize=5)
# Mencetak jumlah elemen saat ini dalam antrean (hasilnya 0 karena baru dibuat)
print(qu.qsize())
# Memasukkan elemen ke dalam antrean (Enqueue)
qu.put('a')
qu.put('i')
qu.put('u')
qu.put('e')
qu.put('o')
print("Full: ", qu.full()) # Mengecek apakah antrean sudah penuh (True/False)
print("size: ", qu.qsize()) # Mencetak jumlah elemen saat ini dalam antrean
# get() function to dequeue element from queue
print("\nElements dequeued from the queue")
# Mengambil dan menghapus elemen pertama yang masuk
print(qu.get())
print(qu.get())
print(qu.get())
print("\nEmpty: ", qu.empty()) # Mengecek apakah antrean kosong
print("Full: ", qu.full()) # Mengecek apakah antrean masih penuh
```

queue dengan single linked list

```python
class Queue: # Mendefinisikan class bernama Queue
  def __init__(self): # Constructor untuk menginisialisasi objek baru
    self.queue = list() # Membuat list kosong bernama 'queue' di dalam objek

  def addtoqu(self,dataval): # menambah elemen ke antrean
    if dataval not in self.queue: # Mengecek apakah data sudah ada (mencegah duplikasi)
      self.queue.insert(0,dataval) # Memasukkan data ke INDEKS 0 (posisi paling depan list)
      return True
    return False

  def removefromq(self): # mengambil elemen dari antrean
    if len(self.queue)>0: # Mengecek apakah ada elemen di dalam list
      return self.queue.pop() # Menghapus dan mengembalikan elemen PALING BELAKANG (pop)
    return ("No elements in Queue!")
  def size(self): # mengecek jumlah elemen dalam antrean
    return len(self.queue)
TheQueue = Queue() # membuat variabel baru berisi queue
# menambahkan ke dalam antrian
print(TheQueue.addtoqu("Jan"))
print(TheQueue.addtoqu("Feb"))
print(TheQueue.addtoqu("March"))
print(TheQueue.addtoqu("April"))
# Mencetak ukuran antrean saat ini
print(TheQueue.size())
# Menghapus elemen
print(TheQueue.removefromq())
print(TheQueue.removefromq())
# Mencetak ukuran antrean saat ini
print(TheQueue.size())
```
