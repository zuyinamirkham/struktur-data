```python
class Node: # membuat node/ 1 kotak data
  def __init__(self, dataval=None): # menyimpan data yang isi default nya none
    self.dataval = dataval # menyimpan data dalam node/isi nya
    self.nextval = None # alamat selanjutnya kosong
#membuat node
n1 = Node("Januari")
n2 = Node("Februari")
n3 = Node("Maret")

class LinkedList: # mengatur kumpulan node
  def __init__(self): # menyimpan data
    self.headval = None # node pertama bernilai none
#Print the linked list
  def listprint(self): # fungsi untuk mencetak semua data
    printval = self.headval # printival: variabel untuk menelusuri tiap node
    while printval is not None: # perulangan sampai variabel tersebut bernilai none
      print (printval.dataval) # menampilkan node saat ini
      printval = printval.nextval # bergerak ke node selanjutnya

  def AddBegining(self, newdata): # menambahkan node baru ke depan linked list
    NewNode = Node(newdata) # variabel yang menunjuk pada node baru
    NewNode.nextval = self.headval # setelah node baru adalah node pertama
    self.headval = NewNode # node pertama berisi node baru
Li = LinkedList() # variabel yang berisi linked list
Li.headval = n1 # pada node pertama berini n1

Li.headval.nextval = n2 # menunjuk pada node kedua yaitu n2

n2.nextval = n3 #menunjuk pada node ketiga yaitu n3
Li.AddBegining("Start") # isi dari node baru
Li.listprint() # program
```

menghapus node dengan kata kunci

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None
class LinkedList:
  def __init__(self):
    self.headval = None

  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

  def AddBegining(self, newdata):
    NewNode = Node (newdata)

    NewNode.nextval = self.headval
    self.headval = NewNode
# menghapus node dengan kata kunci
  def RemoveNode(self, Removekey):
    HeadVal = self.headval # mulai dari head
    if (HeadVal is not None): # kondisi jika head tidak kosong
      if (HeadVal.dataval == Removekey): # kondisi 2 jika head samadengan key
        # head dipindahkan ke node berikutnya
        self.headval = HeadVal.next
        HeadVal = None
        return
    while (HeadVal is not None): # menelusuri tiap node
      if HeadVal.dataval == Removekey: # kondisi jika data di temukan
        break
      prev = HeadVal # simpan node sebelumnya
      HeadVal = HeadVal.nextval # pindah ke node berikutnya
    if (HeadVal == None): # kondisi jika data tidak di temukan
      return
    prev.nextval = HeadVal.nextval # lewati node yang di hapus
    HeadVal = None

Li = LinkedList()

Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Febuari")
print("Sebelum dilakukan remove")
Li.listprint()
print("===========================================")

Li.RemoveNode("Maret") # nama node yang akan di hapus
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
```

hapus node di akhir

```python
class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None
class LinkedList:
  def __init__(self):
    self.headval = None

  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval
# menambahkan di akhir
  def AddInEnd(self, newdata):
    NewNode = Node (newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while(last.nextval):
      last = last.nextval
    last.nextval=NewNode
# menghapus diakhir
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None


Li = LinkedList()

Li.AddInEnd("Januari")
Li.AddInEnd("Februari")
Li.AddInEnd("Maret")
print("Sebelum dilakukan remove")
Li.listprint()
print("============================================")

Li.RemoveEnd() # menghapus di akhir
print("Setelah dilakukan remove node terakhir")
Li.listprint()
```

1.Buatlah double linked list dengan 3 node didalamnya, kemudian lakukan penyisipan tengah dan hapus dengan kata kunci !

```python
class Node:
    def __init__(self, data):
        self.dataval = data
        self.nextval = None
        self.prevval = None

class DoublyLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printival = self.headval
        while printival is not None:
          print(printival.dataval)
          printival = printival.nextval

    # Menyisipkan node di tengah
    def AddInbetween(self, mid_node, newdata):
        if mid_node is None:
            print("Node tidak ditemukan")
            return
        NewNode = Node(newdata)
        NewNode.nextval = mid_node.nextval
        NewNode.prevval = mid_node
        if mid_node.nextval is not None:
            mid_node.nextval.prevval = NewNode
        mid_node.nextval = NewNode

    # Menghapus node berdasarkan kata kunci
    def RemoveNode(self, Removekey):
        HeadVal = self.headval
        if (HeadVal is not None):
            if (HeadVal.dataval == Removekey):
                self.headval = HeadVal.nextval
                if self.headval is not None:
                    self.headval.prevval = None
                HeadVal = None
                return

        while (HeadVal is not None):
            if HeadVal.dataval == Removekey:
                break
            prev = HeadVal
            HeadVal = HeadVal.nextval
        if (HeadVal == None):
            return
        prev.nextval = HeadVal.nextval
        if HeadVal.nextval is not None:
          HeadVal.nextval.prevval = prev
        HeadVal = None

# Membuat double linked list dan 3 node awal
dll = DoublyLinkedList()
node1 = Node("Januari")
node2 = Node("Februari")
node3 = Node("April")

# Menyambungkan node untuk double linked list
dll.headval = node1
node1.nextval = node2
node2.prevval = node1
node2.nextval = node3
node3.prevval = node2

print("Sebelum penyisipan dan penghapusan:")
dll.listprint()
print("===============================")

print("Setelah penyisipan di tengah:")
dll.AddInbetween(node2, "Maret")
dll.listprint()
print("===============================")

print("Setelah penghapusan dengan kata kunci:")
dll.RemoveNode("Februari")
dll.listprint()
print("===============================")
```

2.Buatlah circular linked list dengan 3 node didalamnya, kemudian lakukan penyisipan akhir dan hapus akhir !

```python
class Node:
    def __init__(self, data):
        self.dataval = data # menyimpan data
        self.nextval = None # pointer ke node selanjutnya

class CircularLinkedList:
    def __init__(self):
        self.headval = None

    # Menampilkan isi circular linked list
    def listprint(self):
        if self.headval is None:
            print("List kosong")
            return
        printival = self.headval
        while True:
          print(printival.dataval)
          printival = printival.nextval
          if printival == self.headval:
              break

    # Penyisipan node di akhir
    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return
        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval
        last.nextval = NewNode # node terakhir menunjuk node baru
        NewNode.nextval = self.headval # node baru menunjuk head

    # Menghapus node terakhir
    def RemoveEnd(self):
        if self.headval is None:
            return
        if self.headval.nextval == self.headval:
            self.headval = None
            return
        prev = None
        last = self.headval
        while last.nextval != self.headval:
            prev = last
            last = last.nextval
        prev.nextval = self.headval # node sebelum terakhir menunjuk head
        last = None # hapus node terakhir


cll = CircularLinkedList()

# Membuat 3 node awal
node1 = Node("Januari")
node2 = Node("Februari")
node3 = Node("April")

# Menyambungkan node untuk circular linked list
cll.headval = node1
node1.nextval = node2
node2.nextval = node3
node3.nextval = node1


print("Sebelum penyisipan dan penghapusan:")
cll.listprint()
print("================================")

# Menyisipkan node di akhir
cll.AddInEnd("Mei")
print("Setelah penyisipan node di akhir:")
cll.listprint()
print("================================")

# Menghapus node terakhir
cll.RemoveEnd()
print("Setelah menghapus node terakhir:")
cll.listprint()
print("================================")
```
